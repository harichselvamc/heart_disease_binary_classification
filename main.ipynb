# Dataset Overview

1. age in years
2. sex ( 1 male : 0 female ) 
3. chest pain type 
4. resting blood pressure
5. serum cholestoral
6. fasting blood sugar
7. resting ecg
8. maximum heart rate acheived
9. exercise induced angina
10. ST depression induced by exercise relative to rest
11. slope of the peak exerice to ST segment
12. number of major vessels(0-3)colored by flourosopy
13. thal ( 3: normal , 6 : fixed defect , 7 : reversable defect )
14. target ( 1 disease , 0 no disease)

```python
#divide the dataset for model 
x=heartdf.drop('target',axis=1)
y=heartdf['target']
```

```python
x.head()
```

```python
y.head()
```

```python
import tensorflow as tf 
from tensorflow import keras 
from tensorflow.keras import layers
```

```python
#defining the sequential model for binary classification
model=keras.Sequential()
model.add(layers.Dense(11,activation='relu',input_dim=13))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
#compile and fit the model 

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model.fit(x,y,validation_split=0.22,epochs=300)
```

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel("loss")
plt.xlabel("no of epochs")
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel("accuracy")
plt.xlabel("no of epochs")
plt.legend(['train','test'],loc='upper left')
plt.show()
```
